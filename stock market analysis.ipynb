```python
import pandas as pd
import matplotlib.pyplot as plt

%matplotlib inline
```

# importing dataset

```python
amd=pd.read_csv("AMD (1).csv",header=0,index_col="Date",parse_dates=True)
```

```python
amd.tail()
```

# closing price on diffrent years

```python
y=amd["Adj Close"]
plt.fill_between(amd.index,y,color="olive")
plt.grid(True)
plt.show()
```

# profit genrated with  time

```python
y=amd["Adj Close"]
#plt.plot(amd.index,y,color="r")
plt.plot(amd.index,amd["Close"]-amd["Open"],color="crimson")
plt.grid(True)
plt.show()
```

# increase in closing price with time

```python
y=amd["Close"]
plt.bar(amd.index,y,color="y")
plt.grid(True)
plt.show()
```

# change in volumes of stock

```python
y=amd["Adj Close"]
#plt.plot(amd.index,y,color="r")
fig=plt.plot(amd.index,amd["Volume"],color="orange")
plt.grid(True)
plt.savefig("stock")
plt.show()
```
